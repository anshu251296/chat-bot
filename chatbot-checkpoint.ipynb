```python
import contractions
data['conv1']=data['conv1'].apply(lambda x: contractions.fix(x))
data['conv2']=data['conv2'].apply(lambda x: contractions.fix(x))
```

```python
def remove_emoji(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
data['conv1']=data['conv1'].apply(lambda x: remove_emoji(x))
data['conv2']=data['conv2'].apply(lambda x: remove_emoji(x))
```

```python
#converting into lowercase
data['conv1']=data['conv1'].apply(lambda x: x.lower())
data['conv2']=data['conv2'].apply(lambda x: x.lower())
# data['person1']=data['person1'].apply(lambda x: x.lower())
# data['person2']=data['person2'].apply(lambda x: x.lower())
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
data['conv1']=data['conv1'].apply(lambda x: word_tokenize(x))
data['conv2']=data['conv2'].apply(lambda x: word_tokenize(x))
data.head()
```

```python
def remove_stopwords(x):
  no_stop_words=[]
  for i in x:
    if i not in list(stopwords.words('english')):
      no_stop_words.append(i)
  return no_stop_words
data['conv1']=data['conv1'].apply(lambda x: remove_stopwords(x))
data['conv2']=data['conv2'].apply(lambda x: remove_stopwords(x))
data.head()
```

```python
from nltk.tag import pos_tag
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
def word_lemmatizer(text):
    lem_text=""
    for i in text:
      lem_text+=WordNetLemmatizer().lemmatize(i)+' '
    return lem_text.strip()
data['conv1'] = data['conv1'].apply(lambda x: word_lemmatizer(x))
data['conv2'] = data['conv2'].apply(lambda x: word_lemmatizer(x))
data.head()
```
